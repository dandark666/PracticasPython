### Representacion grafica de los Datos 

Si a la funcion de trasado se le da una matriz de Datos, usara como cordenadas en el eje vertical , y utilizara el indice de cada punto de datos en el array como la coordenada horizontal

```python
plt.plot([1,2,5,7,7,8,3,1])
plt.show
```

Tambien se puede proporcionar 2 matrices: una para el eje horizontal y otra para el eje vertical, 

```python
plt.plot([-3,-1,0,4,7], [1,4,6,7,8])
plt.show()
```

Pueden modificarse las longitudes de los ejes para que la figura no se vea tan ajustada

```python
plt.plot([-3,-1,0,4,7],[1,4,6,7,8])
plt.axis([-4,4,0,10])# [xmin, xmax,ymin,ymax]
plt.show()
```

Se sigue el mismo procedimiento para pintar una funcion matematica.

```python
import numpy as np
x= np.linspace(-2,2,500)
y = x**2
plt.plot(x,y)
plt.show()
```

Tambien puede modificarse el estilo de la grafica para que contenga mas informacion

```python
plt.plot(x,y)
plt.title("Square Function")
plt.xlabel("Matematico")
plt.ylabel("y = x^2")
plt.grid(True)
plt.show()
```

Pueden superponerce graficas y cambiar el estilo de las lineas 

```python
import numpy as np 
x= np.linspace(-2,2,500)
y= x**2
y2 = x+1
plt.plot(x,y,'b--', x,y2,'g')
plt.show()
```

```python
# Separando en diferentes lineas las funciones 
import numpy as np 
x=np.linspace (-2,2,500)
y= x**2
y2=x+1

plt.title("Square Function")
plt.plot(x,y, 'b--')
plt.plot(x,y2, 'g')
plt.show()
```

```python
# Para poder diferenciar  en tre las funciones, siempre es recomendable añadir una leyenda 

import numpy as np 
x=np.linspace (-2,2,500)
y= x**2
y2=x+1

plt.title("Square Function")
plt.plot(x,y, 'b--',label = "X^2")
plt.plot(x,y2, 'g',label = "x+1")
plt.legend(loc = "best") # la situa en la mejor localizacion 
plt.show()
```

Tambien puede crearce do graficas  que no superpongan. Estas graficas se organizan en un grid y se denomina subplots 

```python
import numpy as pd
x= np.linspace(-2,-2-500)
y = x**2
y2 = x + 1

plt.subplot(1,2,1) # 1rows, 2 columns, 1st subplot
plt.plot (x,y, 'b--')

plt.subplot(1,2,2) # 1rows, 2 columns, 1st subplot
plt.plot (x,y2, 'g')
plt.show()
```

#### Scatter Plots

```python
from numpy.random import rand 
x,y = rand(2,100)
plt.scatter(x,y)
plt.show()
```

```python
from numpy.random import rand
x,y = rand(2, 100)
x2, y2 = rand(2, 100)
plt.scatter(x,y,c='red')
plt.scatter(x2, y2, c='blue')
plt.show()
```

#### Histogramas 

```python
data = [1,1.1,1.8,2.2,3.2,3,3,3,3]
plt.subplot(211) # 2row, 1 columna, 1st subplot
plt.hist(data, bins = 10, rwidth=0.8)

plt.subplot(212) # 2row, 1 columna, 1st subplot
plt.hist(data, bins=[1,1.5,2.5,3], rwidth=0.95)
plt.xlabel("value")
plt.xlabel("Frecuencia")
plt.show()
```
